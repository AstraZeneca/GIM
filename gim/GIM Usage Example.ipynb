```python
import numpy as np
import pandas as pd
import gim
import matplotlib.pyplot as plt
```

#### Load the control and treatment replicates for one sample

```python
df_control_replicates = pd.read_csv("data/control_replicates.csv")
df_treatment_replicates = pd.read_csv("data/treatment_replicates.csv")
```

#### Apply the Gene Interaction as Pixel transformation to obtain a single image-like feature matrix

```python
sample_image = gim.gim_transform(df_control_replicates, df_treatment_replicates)
```

```python
plt.imshow(sample_image, cmap = 'viridis')
```
